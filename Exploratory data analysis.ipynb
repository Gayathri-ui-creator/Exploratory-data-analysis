```python
# Survival rate by Sex and Pclass
```

```python
survival_by_sex = df.groupby('Sex')['Survived'].mean()
survival_by_class = df.groupby('Pclass')['Survived'].mean()
```

```python
# Plot survival rate by Sex
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.barplot(x=survival_by_sex.index, y=survival_by_sex.values)
plt.title('Survival Rate by Sex')
plt.ylabel('Survival Rate')
plt.xlabel('Sex')
```

```python
# Plot survival rate by Pclass
```

```python
plt.subplot(1,2,2)
sns.barplot(x=survival_by_class.index, y=survival_by_class.values)
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.xlabel('Pclass')
```

```python
plt.tight_layout()
plt.show()
```

```python
print('Displayed survival rates by sex and class.')
```

```python
# Survival rate by Embarked
```

```python
survival_by_embarked = df.groupby('Embarked')['Survived'].mean()
```

```python
# Family size feature
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
survival_by_familysize = df.groupby('FamilySize')['Survived'].mean()
```

```python
# Plot survival rate by Embarked
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
sns.barplot(x=survival_by_embarked.index, y=survival_by_embarked.values)
plt.title('Survival Rate by Port of Embarkation')
plt.ylabel('Survival Rate')
plt.xlabel('Embarked')
```

```python
# Plot survival rate by Family Size
```

```python
plt.subplot(1,2,2)
sns.lineplot(x=survival_by_familysize.index, y=survival_by_familysize.values, marker='o')
plt.title('Survival Rate by Family Size')
plt.ylabel('Survival Rate')
plt.xlabel('Family Size')
```

```python
plt.tight_layout()
plt.show()
```

```python
print('Displayed survival rates by port of embarkation and family size.')
```

```python
# 2. Value counts for categorical features
```

```python
print('\
--- Value Counts: Sex ---')
print(df['Sex'].value_counts())
print('\
--- Value Counts: Pclass ---')
print(df['Pclass'].value_counts())
print('\
--- Value Counts: Embarked ---')
print(df['Embarked'].value_counts())
print('\
--- Value Counts: Survived ---')
print(df['Survived'].value_counts())
```

```python
# Visualize pairwise relationships
```

```python
import seaborn as sns
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived', 'FamilySize']], hue='Survived', diag_kind='kde')
plt.suptitle('Pairplot of Numerical Features by Survival', y=1.02)
plt.show()
```

```python
# Correlation heatmap
```

```python
plt.figure(figsize=(8,6))
corr = df[['Age', 'Fare', 'Pclass', 'SibSp', 'Parch', 'FamilySize', 'Survived']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
print('Displayed pairplot and correlation heatmap.')
```
